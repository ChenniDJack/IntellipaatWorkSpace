# Advanced Functionality 

```python
coffee.loc[10,["Units Sold"]] = 10
coffee.loc[10,["revenue"]] = coffee.loc[coffee["price",["Units Sold"]] * coffee.loc[10,["price"]]


coffee["yesterday_revenue"] = coffee["revenue"].shift(2)
coffee["PorL"] = coffee['revenue'] - coffee["yesterday_revenue"]
```
